```python
# Función para extraer los nombres de los géneros
def extract_genre_names(genres_str):
    genres_list = ast.literal_eval(genres_str)
    return [genre['name'] for genre in genres_list]

# Aplicar la función a la columna 'genres'
df_movies['genres_list'] = df_movies['genres'].apply(extract_genre_names)


df_movies['genres_string'] = df_movies['genres_list'].apply(lambda x: ', '.join(x))

df_movies.drop(columns='genres_list', inplace=True)
```

```python
# Función personalizada para normalizar la columna "production_companies"
def normalize_production_companies(value):
    try:
        pattern = r"'name':\s+'(.*?)'"
        matches = re.findall(pattern, value)
        return ', '.join(matches)
    except (TypeError, AttributeError):
        return ''

# Aplicamos la función personalizada a cada elemento de la columna "production_companies"
df_movies['company'] = df_movies['production_companies'].apply(normalize_production_companies)

# Imprimimos el DataFrame con la columna de compañías de producción normalizada
df_movies.head()
```

```python
def normalize_production_countries(value):
    if pd.isna(value) or value == '[]':
        return ''
    try:
        name_pattern = r"'name':\s+'(.*?)'"
        iso_pattern = r"'iso_3166_1':\s+'(.*?)'"
        names = re.findall(name_pattern, value)
        isos = re.findall(iso_pattern, value)
        return ', '.join(f"{name} ({iso})" for name, iso in zip(names, isos)) if names else ''
    except (TypeError, AttributeError):
        return ''

# Aplicamos la función modificada
df_movies['countries_with_iso'] = df_movies['production_countries'].apply(normalize_production_countries)

df_movies.head()
```

```python
def check_date_format(date_string):
    pattern = r'^\d{4}-\d{2}-\d{2}$'
    return bool(re.match(pattern, str(date_string)))

# Aplicar la función a la columna 'release_date'
df_movies['valid_date_format'] = df_movies['release_date'].apply(check_date_format)

# Contar cuántas fechas son válidas y cuántas no
valid_count = df_movies['valid_date_format'].sum()
invalid_count = len(df_movies) - valid_count

print(f"Fechas en formato válido: {valid_count}")
print(f"Fechas en formato inválido: {invalid_count}")

# Si quieres ver las fechas inválidas
if invalid_count > 0:
    print("\nEjemplos de fechas inválidas:")
    invalid_dates = df_movies[~df_movies['valid_date_format']]
    print(invalid_dates[['release_date']].head())
```

```python
def convert_to_valid_date(date_string):
    if pd.isna(date_string):
        return date_string
    
    # Si ya está en el formato correcto, devuélvelo tal cual
    if re.match(r'^\d{4}-\d{2}-\d{2}$', str(date_string)):
        return date_string
    
    # Intenta convertir el string a una fecha
    try:
        date = pd.to_datetime(date_string, errors='raise')
        return date.strftime('%Y-%m-%d')
    except:
        # Si la conversión falla, intenta manejar casos específicos
        if str(date_string).isdigit():
            if len(str(date_string)) == 1 or len(str(date_string)) == 2:
                # Asume que es un día del mes actual
                today = pd.Timestamp.now()
                return f"{today.year}-{today.month:02d}-{int(date_string):02d}"
        
        # Si no se puede convertir, devuelve None o un valor que indique error
        return None

# Aplica la función de conversión solo a las fechas inválidas
mask = ~df_movies['valid_date_format']
df_movies.loc[mask, 'release_date'] = df_movies.loc[mask, 'release_date'].apply(convert_to_valid_date)

# Vuelve a verificar el formato
df_movies['valid_date_format'] = df_movies['release_date'].apply(check_date_format)

# Cuenta nuevamente
valid_count = df_movies['valid_date_format'].sum()
invalid_count = len(df_movies) - valid_count

print(f"Fechas en formato válido después de la conversión: {valid_count}")
print(f"Fechas en formato inválido después de la conversión: {invalid_count}")

# Si aún quedan fechas inválidas, muéstralas
if invalid_count > 0:
    print("\nEjemplos de fechas que siguen siendo inválidas:")
    invalid_dates = df_movies[~df_movies['valid_date_format']]
    print(invalid_dates[['release_date']].head())
```

```python
def extract_actor_names(cast_string):
    # Extraer directamente los nombres de los actores usando una expresión regular
    names = re.findall(r"'name': '(.*?)'", cast_string)
    return ', '.join(names)

# Aplicar la función optimizada directamente para obtener los nombres de los actores
df_creditos['cast_names'] = df_creditos['cast'].apply(extract_actor_names)
```

```python
def extract_directors(crew_string):
    pattern = r"'job': 'Director'.*?'name': '(.*?)'"
    directors = ', '.join(set(re.findall(pattern, crew_string)))
    return directors

# Aplicar la función optimizada
df_creditos['directors'] = df_creditos['crew'].apply(extract_directors)

df_creditos.drop(columns='crew', inplace=True)
```
